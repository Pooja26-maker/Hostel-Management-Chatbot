```python
import streamlit as st
import pandas as pd
from datetime import datetime
import os

st.set_page_config(page_title="Hostel Complaint Bot", page_icon="🏠")

st.title("🏠 Smart Hostel Complaint Bot")

# Complaint rules
rules = {
    "water leakage": {
        "department": "Maintenance Team",
        "priority": "High"
    },
    "wifi issue": {
        "department": "IT Team",
        "priority": "Medium"
    },
    "electricity issue": {
        "department": "Electrical Team",
        "priority": "High"
    },
    "food issue": {
        "department": "Mess Team",
        "priority": "Medium"
    },
    "cleaning issue": {
        "department": "Housekeeping",
        "priority": "Low"
    }
}

# Inputs
student_name = st.text_input("Enter your name")
room_number = st.text_input("Enter room number")

complaint = st.text_area("Enter your complaint")

# Function
def match_complaint(text):
    text = text.lower()

    for key in rules:
        if key in text:
            return (
                key,
                rules[key]["department"],
                rules[key]["priority"]
            )

    return ("Unknown", "Admin Team", "Low")

# Button
if st.button("Submit Complaint"):

    category, department, priority = match_complaint(complaint)

    st.success(f"""
Complaint Registered Successfully!

Category: {category}

Department: {department}

Priority: {priority}
""")

    # Save data
    data = {
        "Time": [datetime.now()],
        "Student": [student_name],
        "Room": [room_number],
        "Complaint": [complaint],
        "Category": [category],
        "Department": [department],
        "Priority": [priority]
    }

    df = pd.DataFrame(data)

    file_name = "complaints.csv"

    if os.path.exists(file_name):
        old_df = pd.read_csv(file_name)
        df = pd.concat([old_df, df], ignore_index=True)

    df.to_csv(file_name, index=False)

    st.info("Complaint saved successfully!")

# Show complaint history
if os.path.exists("complaints.csv"):

    st.subheader("📋 Complaint History")

    history = pd.read_csv("complaints.csv")

    st.dataframe(history)
```

```python
import streamlit as st
import pandas as pd
from datetime import datetime
import os

# Page Config
st.set_page_config(
    page_title="Hostel Complaint Bot",
    page_icon="🏠",
    layout="centered"
)

# Title
st.title("🏠 Hostel Complaint Bot")
st.markdown("Smart AI-style complaint management system")

# Inputs
name = st.text_input("Enter your name")

room = st.text_input("Enter your room number")

complaint = st.text_area("Describe your complaint")


# Smart Complaint Detection
def detect_issue(text):

    text = text.lower()

    if "water" in text or "leak" in text:
        return "Water Problem", "Maintenance Team", "High"

    elif "wifi" in text or "internet" in text:
        return "WiFi Issue", "IT Team", "Medium"

    elif "electricity" in text or "fan" in text or "light" in text:
        return "Electrical Problem", "Electrical Team", "High"

    elif "food" in text or "mess" in text:
        return "Food Complaint", "Mess Team", "Medium"

    elif "clean" in text or "dirty" in text:
        return "Cleaning Issue", "Housekeeping", "Low"

    else:
        return "General Complaint", "Admin Team", "Low"


# Submit Button
if st.button("Submit Complaint"):

    category, department, priority = detect_issue(complaint)

    st.success("Complaint Submitted Successfully!")

    st.write("## Complaint Details")
    st.write(f"👤 Student Name: {name}")
    st.write(f"🏠 Room Number: {room}")
    st.write(f"📝 Complaint: {complaint}")
    st.write(f"🏷 Category: {category}")
    st.write(f"🏢 Department: {department}")
    st.write(f"⚡ Priority: {priority}")

    # Save complaint to CSV
    data = {
        "Time": [datetime.now()],
        "Student Name": [name],
        "Room Number": [room],
        "Complaint": [complaint],
        "Category": [category],
        "Department": [department],
        "Priority": [priority]
    }

    df = pd.DataFrame(data)

    file_name = "complaints.csv"

    if os.path.exists(file_name):
        old_df = pd.read_csv(file_name)
        df = pd.concat([old_df, df], ignore_index=True)

    df.to_csv(file_name, index=False)

    st.info("Complaint saved successfully!")


# Show Complaint History
if os.path.exists("complaints.csv"):

    st.write("## 📋 Complaint History")

    history = pd.read_csv("complaints.csv")

    st.dataframe(history)
```

```python
import streamlit as st
import pandas as pd
from datetime import datetime
import os
import matplotlib.pyplot as plt

# Page Config
st.set_page_config(
    page_title="Hostel Complaint Management System",
    page_icon="🏠",
    layout="wide"
)

# Sidebar
st.sidebar.title("🏠 Hostel Complaint System")

menu = st.sidebar.radio(
    "Navigation",
    ["Submit Complaint", "Admin Dashboard"]
)

# File name
FILE_NAME = "complaints.csv"

# Smart complaint detection
def detect_issue(text):

    text = text.lower()

    if "water" in text or "leak" in text:
        return "Water Problem", "Maintenance Team", "High"

    elif "wifi" in text or "internet" in text:
        return "WiFi Issue", "IT Team", "Medium"

    elif "electricity" in text or "fan" in text or "light" in text:
        return "Electrical Problem", "Electrical Team", "High"

    elif "food" in text or "mess" in text:
        return "Food Complaint", "Mess Team", "Medium"

    elif "clean" in text or "dirty" in text:
        return "Cleaning Issue", "Housekeeping", "Low"

    else:
        return "General Complaint", "Admin Team", "Low"


# -------------------------------
# Submit Complaint Page
# -------------------------------

if menu == "Submit Complaint":

    st.title("🏠 Hostel Complaint Bot")

    st.markdown("AI-powered complaint triage system")

    name = st.text_input("Enter your name")

    room = st.text_input("Enter room number")

    complaint = st.text_area("Describe your complaint")

    if st.button("Submit Complaint"):

        category, department, priority = detect_issue(complaint)

        st.success("Complaint Submitted Successfully!")

        st.write("## Complaint Details")
        st.write(f"👤 Student Name: {name}")
        st.write(f"🏠 Room Number: {room}")
        st.write(f"📝 Complaint: {complaint}")
        st.write(f"🏷 Category: {category}")
        st.write(f"🏢 Department: {department}")
        st.write(f"⚡ Priority: {priority}")

        # Save data
        data = {
            "Time": [datetime.now()],
            "Student": [name],
            "Room": [room],
            "Complaint": [complaint],
            "Category": [category],
            "Department": [department],
            "Priority": [priority],
            "Status": ["Pending"]
        }

        df = pd.DataFrame(data)

        if os.path.exists(FILE_NAME):
            old_df = pd.read_csv(FILE_NAME)
            df = pd.concat([old_df, df], ignore_index=True)

        df.to_csv(FILE_NAME, index=False)

        st.info("Complaint stored successfully!")


# -------------------------------
# Admin Dashboard
# -------------------------------

elif menu == "Admin Dashboard":

    st.title("📊 Admin Dashboard")

    if os.path.exists(FILE_NAME):

        df = pd.read_csv(FILE_NAME)

        # Metrics
        total = len(df)

        high = len(df[df["Priority"] == "High"])

        pending = len(df[df["Status"] == "Pending"])

        col1, col2, col3 = st.columns(3)

        col1.metric("Total Complaints", total)

        col2.metric("High Priority", high)

        col3.metric("Pending Complaints", pending)

        st.write("---")

        # Search
        search = st.text_input("🔍 Search Complaint")

        if search:
            filtered_df = df[
                df["Complaint"].str.contains(search, case=False)
            ]
        else:
            filtered_df = df

        st.write("## Complaint Records")

        st.dataframe(filtered_df)

        st.write("---")

        # Charts
        st.write("## 📈 Complaint Analytics")

        category_counts = df["Category"].value_counts()

        fig, ax = plt.subplots()

        ax.pie(
            category_counts,
            labels=category_counts.index,
            autopct='%1.1f%%'
        )

        st.pyplot(fig)

    else:
        st.warning("No complaints found!")
```
